* Function that replaces the outliers with median values

```python
def get_median(df, model):
    model_median = df[df['MODEL'] == model]['PRICE'].median()
    for index, row in df.iterrows():
        price = row['PRICE']
        mod = row['MODEL']
        if price > 400000.0 and mod == model:
            df.at[index, 'PRICE'] = model_median
```

**#Q5:** *Are cars with automatic transmissions typically more expensive than cars with manual transmissions?*

```python
automatic_mean = df.loc[df['TRANSMISSION'] == 'automatic']['PRICE'].mean()
manual_mean = df.loc[df['TRANSMISSION'] == 'manual']['PRICE'].mean()

manual_vs_automatic = {
    'TRANS' : ['Automatic', 'Manual'],
    'MEAN_PRICES' : [automatic_mean, manual_mean]
}
manual_vs_automatic_df = pd.DataFrame(manual_vs_automatic)
```

```python
plt.bar(manual_vs_automatic_df['TRANS'] , manual_vs_automatic_df['MEAN_PRICES'], edgecolor = 'Black', width = 0.5, color = ['navy', 'coral'])

plt.title('Average Prices Between Automatic & Manual Transmissions', fontweight = 'bold')
plt.xlabel('Transmission')
plt.ylabel('Average Prices')
```
